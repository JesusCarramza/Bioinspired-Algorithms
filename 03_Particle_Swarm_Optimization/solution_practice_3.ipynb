```python
import numpy as np

# 1. Parámetros
N = 20
iteraciones = 50
a = 0.4
b1 = 0.7
b2 = 1.2

# 2. Función objetivo
def funcion_objetivo(x, y):
    return x**2 + y**2 + 25 * (np.sin(x) + np.sin(y))

# 3. Inicialización
posiciones = np.random.uniform(-5, 5, (N, 2))
velocidades = np.zeros((N, 2))

# pbest inicial = posición inicial
pbest_pos = np.copy(posiciones)
pbest_val = np.array([funcion_objetivo(p[0], p[1]) for p in posiciones])

# gbest inicial
gbest_idx = np.argmin(pbest_val)
gbest_pos = np.copy(posiciones[gbest_idx])
gbest_val = pbest_val[gbest_idx]

print("=== Inicio de la Optimización PSO ===")

# Ciclo Principal
for t in range(iteraciones):
    print(f"\n--- Iteración {t+1} ---")

    # Guardar fitness de esta iteración
    fitness_iteracion = np.zeros(N)
    posiciones_iteracion = np.zeros((N, 2))

    for i in range(N):

        r1 = np.random.uniform(0, 1)
        r2 = np.random.uniform(0, 1)

        # Actualizar velocidad
        velocidades[i] = (
            a * velocidades[i]
            + b1 * r1 * (pbest_pos[i] - posiciones[i])
            + b2 * r2 * (gbest_pos - posiciones[i])
        )

        # Actualizar posición
        posiciones[i] = posiciones[i] + velocidades[i]
        posiciones[i] = np.clip(posiciones[i], -5, 5)

        # Evaluar fitness actual
        aptitud_actual = funcion_objetivo(posiciones[i][0], posiciones[i][1])

        # Nuevo pbest: Comparacon el pbest anterior
        if aptitud_actual < pbest_val[i]:
            nuevo_pbest_val = aptitud_actual
            nuevo_pbest_pos = np.copy(posiciones[i])
        else:
            nuevo_pbest_val = pbest_val[i]
            nuevo_pbest_pos = np.copy(pbest_pos[i])

        # Guardar nuevo pbest para próxima iteración
        pbest_val[i] = nuevo_pbest_val
        pbest_pos[i] = nuevo_pbest_pos

        # Guardar datos de esta iteración para gbest
        fitness_iteracion[i] = aptitud_actual
        posiciones_iteracion[i] = np.copy(posiciones[i])

    # Nuevo gbest: solo mejor partícula de esta iteración
    idx_mejor_iter = np.argmin(fitness_iteracion)
    gbest_val = fitness_iteracion[idx_mejor_iter]
    gbest_pos = np.copy(posiciones_iteracion[idx_mejor_iter])

    # Imprimir resultados
    for i in range(N):
        print(f"Partícula {i+1:02d} | "
                f"Posición: [{posiciones[i][0]:7.4f}, {posiciones[i][1]:7.4f}] | "
                f"Velocidad: [{velocidades[i][0]:7.4f}, {velocidades[i][1]:7.4f}] | "
                f"pbest: {pbest_val[i]:7.4f} | "
                f"gbest: {gbest_val:7.4f}")

print("\n=== Resultado Final ===")
print(f"Mejor valor encontrado: {gbest_val:.4f}")
print(f"Coordenadas: X={gbest_pos[0]:.4f}, Y={gbest_pos[1]:.4f}")
```
